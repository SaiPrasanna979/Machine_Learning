```python
x=pd.read_csv('hey.csv')
```

```python
x.nunique()
```

```python
x=x[['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']]
```

```python
x.head(
)
```

```python
x.shape
```

```python
x.isnull().sum()
```

```python
#Handling Missong values
```

```python
x[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']]=x[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']].fillna(0)
```

```python
x.isnull().sum()
```

```python
x.columns
```

```python
x['Landsize']=x['Landsize'].fillna(x['Landsize'].mean())
```

```python
x['BuildingArea']=x['BuildingArea'].fillna(x['BuildingArea'].mean())
```

```python
x.dropna(inplace=True)
```

```python
x.isnull().sum()
```

```python
x.shape
```

```python
x.head()
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
X=x.drop('Price',axis=1)
```

```python
y=x['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m=LinearRegression()
```

```python
m.fit(x_train,y_train)
```

```python
m.score(x_test,y_test)
```

```python
m.score(x_train,y_train)
```

```python
from sklearn import linear_model
```

```python
#Lasso L1 Regualarization
model=linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
#Ridge L2 Regualarization
```

```python
model=linear_model.Ridge(alpha=50, max_iter=100, tol=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m=KNeighborsClassifier(n_neighbors=10).fit(X_train,y_train)
```

```python
m.score(X_test,y_test)
```

```python
m.predict([[6.3,3.3,6.0,2.5]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=m.predict(X_test)
```

```python
cm=confusion_matrix(y_pred,y_test)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('Pred')
plt.ylabel('Truth')
plt.figure(figsize=(10,9))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
